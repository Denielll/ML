```python
# unsupervised learning
```

```python
# How it works

# 1, randomly assign each point to K groups
# 2, find the centroid (質心) of each group
# 3, assign each point to the closest centroid 
# repeat 2,3 until the cluster variation can't be reduced any further
```

```python
# not right or wrong about the K value. Elbow Method can be used to choose the K value, also Domain Knowledge

# https://stackoverflow.com/questions/6645895/calculating-the-percentage-of-variance-measure-for-k-means
```

```python
#  not suitable for data with categorical variables !!!  Please refer to K-modes 
```

# make tool data

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
data[0].shape  
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap="rainbow")   
```

# K-Means

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
# compare the kmeans result & original data labels

fig, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title("K Means")
ax1.scatter(data[0][:,0], data[0][:,1], c = kmeans.labels_, cmap="rainbow")

ax2.set_title("Original")
ax2.scatter(data[0][:,0], data[0][:,1], c = data[1], cmap="rainbow" )
```
